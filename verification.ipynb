```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erf
```

```python
D=0.05
U=1
alpha = D/U
num_curves = 11
z_values = np.linspace(0, 1, num_curves) 
x = np.linspace(0, 1, 100) 

    
plt.figure(figsize=(8, 6))
cmap = plt.get_cmap('viridis')
for z in z_values:
    delta = erf(x / np.sqrt(4 * alpha * z))
    color = cmap(z)
    plt.plot(x, delta, label='z = {}'.format("{:.2f}".format(z)), color=color)

plt.xlabel('x', fontsize=18)
plt.ylabel(r'$\delta$', fontsize=18)
plt.yticks(fontsize=16)
plt.xticks(fontsize=16)

plt.legend(fancybox=True, shadow=True, prop={'size': 15})


plt.grid()
plt.show()
```

```python
def erffunc(x,x0,alpha,z,c1,c2):
    return 0.5*(c1-c2)*erf((x-x0)/np.sqrt(4 * alpha * z))
```

```python
D=0.05
U=1
alpha = D/U
num_curves = 11
z_values = np.linspace(0, 1, num_curves) 
x = np.linspace(0, 1, 20) 

    
plt.figure(figsize=(8, 6))
cmap = plt.get_cmap('viridis')
for z in z_values:
    delta = erffunc(x,0.5,alpha,z,-0.01,0.01)
    color = cmap(z)
    plt.scatter(x, delta, label='z = {}'.format("{:.2f}".format(z)), color=color)

plt.xlabel('x', fontsize=18)
plt.ylabel(r'$\delta$', fontsize=18)
plt.yticks(fontsize=16)
plt.xticks(fontsize=16)

plt.legend(fancybox=True, shadow=True, prop={'size': 15})


plt.grid()
plt.show()
```
